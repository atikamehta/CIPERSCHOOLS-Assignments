### LOADING THE  ADULT  DATASET:

###### IMPORTING THE USEFUL LIBRARIES:

```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
from sklearn import preprocessing
import warnings
warnings.filterwarnings('ignore')
```

###### LOADING THE DATASET:

```python
missing_val=['?','-']
df=pd.read_csv('adult.csv',na_values=missing_val)
df
```

###### DELETING THE COLUMN "fnlwgt":

```python
del df['fnlwgt']
df.head()
```

###### DROPING ALL THE ROWS WITH MISSING VALUES:

```python
df=df.dropna(how='any')
df.head()
```

###### AFTER REMOVING UNECCESSORY COLUMNS AND ROWS THE SHAPE IS:

```python
new_a=df.shape[0]
print('the rows left are:',new_a)
print('the columns are :',df.shape[1])
```

###### PERCENTAGE OF MISSING VALUES IN DATASET:

```python
b=a-new_a
per=(b/a)*100
new_per=per*10
print('percentage:',new_per)
```

```python
df.info()
```

###### REMOVING COLUMNS WITH EXCESS OF OUTLIERS:

```python
df=df.drop(['capital-gain','capital-loss'],axis=1)
df.head()
```

###### MAKING NEW DATAFRAME FOR ONLY NUMERICAL COLUMNS:

```python
df_num=df.select_dtypes(include=['int64'])
df_num.head()
```

###### MAKING NEW DATAFRAME FOR ONLY CATEGORICAL COLUMNS:

```python
df_cat=df.select_dtypes(include=['object'])
df_cat.head()
```

###### USING MIN MAX SCALER FOR NORMALIZING DATA:

```python
x=df_num['age'].values.reshape(-1,1)
min_max_scaler=preprocessing.MinMaxScaler()
x_age=min_max_scaler.fit_transform(x)
df_num['age']=x_age

x=df_num['educational-num'].values.reshape(-1,1)
min_max_scaler=preprocessing.MinMaxScaler()
x_edu=min_max_scaler.fit_transform(x)
df_num['educational-num']=x_edu

x=df_num['hours-per-week'].values.reshape(-1,1)
min_max_scaler=preprocessing.MinMaxScaler()
x_hours=min_max_scaler.fit_transform(x)
df_num['hours-per-week']=x_hours

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

df_num=pd.DataFrame(scaler.fit_transform(df_num),columns=df_num.columns)
```

```python
print(df_num.shape)
df_num.head()
```

###### USING DUMMY VARIABLES FOR STANDARDIZING DATA:

```python
df_cat=pd.get_dummies(df_cat,drop_first=True)
df_cat.head()
```

###### JOINING BOTH DATAFRAMES TOGETHER:

```python
df=pd.concat([df_num,df_cat],axis=1)
df.head()
```

```python
df=df.dropna(how='any')
```

```python
df.shape
```

```python
X=df
y=df.pop('income_>50K')
```

###### IMPORTING TRAIN TEST SPLIT:

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=100)
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

###### LOGISTIC REGRESSION:

```python
from sklearn.linear_model import LogisticRegression 
from sklearn import metrics
```

```python
reg=LogisticRegression()
reg.fit(X_train,y_train)
y_train_predict=reg.predict(X_train)
y_test_predict=reg.predict(X_test)
```

```python
print('accuracy score:',metrics.accuracy_score(y_test,y_test_predict))
print('confusion matrix: \n',metrics.confusion_matrix(y_test,y_test_predict))
print('recall score:',metrics.recall_score(y_test,y_test_predict))
print('precision score:',metrics.precision_score(y_test,y_test_predict))
```

###### CLASSIFICATION REPORT FOR LOGISTIC REGRESSION:

```python
print(metrics.classification_report(y_test,y_test_predict))
#number of features = 94
```

######  IMPORTING PCA:

```python
from sklearn.decomposition import PCA
pca=PCA(random_state =0)
pca.fit(X_train)
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
pca_final=PCA(n_components=20,random_state=41)

X_train_pca=pca_final.fit_transform(X_train)
X_test_pca=pca_final.transform(X_test)
```

```python
from sklearn import metrics
reg_pca=LogisticRegression()
reg_pca.fit(X_train_pca,y_train)
y_test_predict_pca=reg_pca.predict(X_test_pca)
```

```python
from sklearn import metrics
print('accuracy score:',metrics.accuracy_score(y_test,y_test_predict_pca))
print('confusion matrix: \n',metrics.confusion_matrix(y_test,y_test_predict_pca))
print('recall score:',metrics.recall_score(y_test,y_test_predict_pca))
print('precision score:',metrics.precision_score(y_test,y_test_predict_pca))
```

###### CLASSIFICATION REPORT FOR PCA:

```python
print(metrics.classification_report(y_test,y_test_predict_pca))
```
